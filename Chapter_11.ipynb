##Map and Transform

```python
tips_data_pd = pd.read_csv("tips.csv")
tips_data_pd.head(5)
```

```python
def double_value(x):
  return x*2
```

```python
tips_data_pd['total_bill'].map(double_value)
```

```python
tips_data_pd.transform(lambda x:x*2)
```

##Merging and Joining

```python
df1 = pd.DataFrame({"key":["A","B","C"],"value1":[1,2,3]})
df2 = pd.DataFrame({"key":["A","B","C"],"value2":[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.concat([df1,df2],axis=1)
```

##Exercise 11.2 Second PandasPolars Exercise

```python
url2 = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv"
horse = pd.read_csv(url2, header=None, na_values="?")
horse.head(5)
```

```python
horse = horse.interpolate(method='linear', limit_direction='both')
horse.head(5)
```

```python
horse["new_column"] = horse[0].map(lambda x:x+10)
horse.head(5)
```

##Exercise 11.3 Third Pandas & Polars Exercise

```python
url3 = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

airline = pd.read_csv(url3)

airline.head(5)
```

```python
airline["Year"] = airline["Month"].str.split('-').str[0]
airline["Month"] = airline["Month"].str.split('-').str[1]
airline.head(5)
```

```python
airline.isnull().sum()
```

```python
airline["Adjusted Passengers"]=airline["Passengers"].apply(lambda x:x+10)
airline.head(5)
```
